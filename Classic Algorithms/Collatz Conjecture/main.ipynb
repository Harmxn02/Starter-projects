```python
from sys import path
import pandas as pd


# import the function 'collatz' from main.py

path.append('..')
from main import collatz
```

```python
NUMBER = 200
steps = collatz(NUMBER)


df = pd.DataFrame({"step": range(1, len(steps)+1), "number": steps})
```

It always ends on 16-8-4-2-1

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

sns.set(style="whitegrid")


# plot the data
plt.figure(figsize=(16, 9))
sns.lineplot(df, x="step", y="number", linewidth=2, dashes=False, marker="o", markersize=10)

# set the Y-axis to always show full numbers
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))

# titles and lines
plt.suptitle(f"Collatz Conjecture on n={NUMBER}", fontweight="bold", y=0.935)
plt.title(f"steps to get to one: {len(steps)}", y=1)
plt.axhline(1, linestyle="--", color="red")

# final touches
sns.despine()
plt.show()
```
